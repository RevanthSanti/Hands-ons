```python
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt


# Perform PCA for 3 components
pca = decomposition.PCA(n_components=3)
X_centered = X - X.mean(axis=0)
pca.fit(X_centered)
X_pca = pca.transform(X_centered)

# print(X_pca)

# Plotting the results of PCA in 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# We plot each class with a different color and label
ax.scatter(X_pca[y == 0, 0], X_pca[y == 0, 1], X_pca[y == 0, 2], color='blue', label='Setosa')
ax.scatter(X_pca[y == 1, 0], X_pca[y == 1, 1], X_pca[y == 1, 2], color='green', label='Versicolour')
ax.scatter(X_pca[y == 2, 0], X_pca[y == 2, 1], X_pca[y == 2, 2], color='red', label='Virginica')

# Setting the labels for each axis
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')

# Adding a legend to identify each class
ax.legend(loc='best')

# Display the plot
plt.show()

```

### Why PCA is used:

**Dimensionality Reduction:** The most common use of PCA is to reduce the dimensionality of data while retaining as much variance as possible.

**Noise Reduction:** By focusing on the components with higher variance, PCA can help to filter out noise from the data.

**Visualization**: Reducing a high-dimensional dataset to two or three principal components can allow for visual representation, which can help identify patterns or groups in the data.

**Feature Extraction and Engineering:** PCA can be used to derive new features that are more informative and less redundant than the original data.

**Avoid Overfitting**: Reducing the number of features through PCA can lead to simpler models that are less likely to overfit.

## Using t-SNE

Handwritten numbers dataset have 0 to 9 digits. The numbers are represented by 8 x 8 matrixes with the color intensity for each pixel. Every matrix is flattened into a vector of 64 numbers, so we get the feature version of the data.

```python
dig_data = pd.read_csv('dataset2.csv')
X = dig_data.drop(columns='target')
y = dig_data['target']
```

Our data has 64 dimensions, but we are going to reduce it to only 2 and see that, even with just 2 dimensions, we can clearly see that digits separate into clusters.

```python
pca = decomposition.PCA(n_components=2)
X_reduced = pca.fit_transform(X)

print('Projecting %d-dimensional data to 2D' % X.shape[1])

plt.figure(figsize=(12,10))
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y,
            edgecolor='none', alpha=0.7, s=40,
            cmap=plt.cm.get_cmap('nipy_spectral', 10))
plt.colorbar()
plt.title('MNIST. PCA projection');
```

Indeed, with t-SNE, the picture looks better since PCA has a linear constraint while t-SNE does not. However, even with such a small dataset, the t-SNE algorithm takes significantly more time to complete than PCA.

```python
from sklearn.manifold import TSNE
tsne = TSNE(random_state=17)

X_tsne = tsne.fit_transform(X)

plt.figure(figsize=(12,10))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y,
            edgecolor='none', alpha=0.7, s=40,
            cmap=plt.cm.get_cmap('nipy_spectral', 10))
plt.colorbar()
plt.title('MNIST. t-SNE projection');
```

```python
X = df.drop(['CustomerID', 'Gender'], axis=1) ## We will play with 3 datapoints
```

```python
from sklearn.cluster import KMeans

clusters = []
plt.clf()
for i in range(1, 11):
    km = KMeans(n_clusters=i).fit(X)
    clusters.append(km.inertia_)

fig, ax = plt.subplots(figsize=(12, 8))
sns.lineplot(x=list(range(1, 11)), y=clusters, ax=ax)
ax.set_title('Searching for Elbow')
ax.set_xlabel('Clusters')
ax.set_ylabel('Inertia')

plt.show()
```

* Elbow method tells us to select the cluster when there is a significant change in inertia.
* As we can see from the graph, we can say this may be either 3 or 5. Let's see both results in graph and decide.

```python
display(X)
```

```python
# 3 cluster
km3 = KMeans(n_clusters=3).fit(X)

X['Labels'] = km3.labels_
plt.figure(figsize=(12, 8))
sns.scatterplot(x=X['Income'], y=X['Score'], hue=X['Labels'],
                palette=sns.color_palette('hls', 3))
plt.title('KMeans with 3 Clusters')
plt.show()
```

```python
# Let's see with 5 Clusters
km5 = KMeans(n_clusters=5).fit(X)

X['Labels'] = km5.labels_
plt.figure(figsize=(12, 8))
sns.scatterplot(x=X['Income'], y=X['Score'], hue=X['Labels'],
                palette=sns.color_palette('hls', 5))
plt.title('KMeans with 5 Clusters')
plt.show()
```

* By judging from the plots we could say that 5 cluster seems better than the 3 ones.

* We can analyze our 5 clusters in detail now:

    * Label 0 is low income and low spending
    * Label 1 is high income and high spending
    * Label 2 is mid income and mid spending
    * Label 3 is high income and low spending
    * Label 4 is low income and high spending
